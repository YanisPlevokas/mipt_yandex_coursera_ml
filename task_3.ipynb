```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создадим выборку с помощью Гамма-распределения.
https://en.wikipedia.org/wiki/Gamma_distribution

```python
a = 2.5
sample = sts.gamma.rvs(a,size = 1000)
```

Нарисуем гистограмму для наших значений.

```python
plt.hist(sample, density = True, color = 'red')
plt.plot(sample, sts.gamma.pdf(a, sample))
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

Создадим двумерный массив, где в строках будут лежать выборочные средние. В первой строке - выборочные средние для выборок из 5 элементов, во второй - из 10, в третьей - из 15.

```python
z = 0
mas = np.zeros((3,1000))
for n in [5, 10, 50]:
    for i in range(1000):
        sample = sts.gamma.rvs(a,size = n)
        mas[z][i] = sum(sample)/n
    z = z + 1
mas
```

```python
plt.hist(mas[0], density = True, color = 'red')
plt.ylabel('fraction of samples')
plt.xlabel('$x, n = 10$')
```

```python
plt.hist(mas[1], density = True, color = 'blue')
plt.ylabel('fraction of samples')
plt.xlabel('$x, n = 10$')
```

```python
plt.hist(mas[2], density = True, color = 'green')
plt.ylabel('fraction of samples')
plt.xlabel('$x, n = 50$')
```

Посчитаем значения параметров нормальных распределений при помощи теоретических мат.ожидания и дисперсии

```python
natur_first_param_5 = sts.gamma.mean(a)
natur_second_param_5 = sts.gamma.var(a)/5
```

Вычислим параметры для выборки размера 5.

```python
plt.hist(mas[0], density = True, color = 'red')
norm_distribution = sts.norm(natur_first_param_5, np.sqrt(natur_second_param_5))
new_sample = norm_distribution.rvs(1000)
x = np.linspace(-4,4, 1000)
pdf = norm_distribution.pdf(x)
plt.plot(x, pdf)
plt.ylabel('f(x)')
plt.xlabel('x')
```

Гистограмма и плотность распределения для выборки с объемом 5.
